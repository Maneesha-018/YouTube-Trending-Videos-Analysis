```python
plt.figure()
plt.bar(df['video_id'].head(10), df['views'].head(10))

plt.xticks(rotation=90)
plt.title("Views of First 10 Videos")
plt.show()
```

```python
top_views = df.sort_values(by='views', ascending=False).head(10)

plt.figure()
plt.bar(range(len(top_views)), top_views['views'])

plt.xticks(range(len(top_views)), range(1,11))  # simple numbers
plt.xlabel("Top Videos")
plt.ylabel("Views")
plt.title("Top 10 Videos by Views")

plt.show()
```

```python
#like vs views
plt.figure()
plt.scatter(df['views'], df['likes'])
plt.xlabel("Views")
plt.ylabel("Likes")
plt.title("Views vs Likes")
plt.show()
```

```python
#like vs dislikes
plt.figure()
plt.scatter(df['likes'], df['dislikes'])
plt.xlabel("Likes")
plt.ylabel("Dislikes")
plt.title("Likes vs Dislikes")
plt.show()
```

```python
#comment distribution 
plt.figure()
plt.hist(df['comment_count'], bins=10)
plt.xlabel("Comment Count")
plt.ylabel("Frequency")
plt.title("Distribution of Comments")
plt.show()
```

```python
#Category wise video count

category_count = df['category_id'].value_counts()

plt.figure()
plt.bar(category_count.index, category_count.values)
plt.xlabel("Category ID")
plt.ylabel("Number of Videos")
plt.title("Videos per Category")
plt.show()
```

```python
# Top Channels
top_channels = df['channel_title'].value_counts().head(10)

plt.figure(figsize=(10,6))

# Clean text (important for avoiding glyph warnings)
channels = top_channels.index.astype(str)
channels = [c.encode('ascii', errors='ignore').decode('ascii') for c in channels]

plt.barh(channels, top_channels.values)

plt.xlabel("Number of Videos")
plt.ylabel("Channel Name")
plt.title("Top 10 Channels with Most Trending Videos")

plt.gca().invert_yaxis()  # highest on top

plt.tight_layout()
plt.show()
```

```python
#engagement Ratio
df['engagement'] = df['likes'] / df['views']

plt.figure()
plt.hist(df['engagement'], bins=50)
plt.xlabel("Engagement Ratio")
plt.ylabel("Frequency")
plt.title("Engagement Distribution")
plt.show()
```

```python
#Correelation Heatmap
import numpy as np

corr = df[['views','likes','dislikes','comment_count']].corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")
plt.show()
```

```python
#Category Distribution
category_count = df['category_id'].value_counts().head(6)

plt.figure()
plt.pie(category_count.values, labels=category_count.index, autopct='%1.1f%%')
plt.title("Top Categories Distribution")
plt.show()
```

```python
#Category Distribution
category_count = df['category_id'].value_counts().head(6)

plt.figure()
plt.pie(category_count.values, labels=category_count.index, autopct='%1.1f%%')

# Donut hole
centre_circle = plt.Circle((0,0),0.70,fc='white')
plt.gca().add_artist(centre_circle)

plt.title("Donut Chart - Categories")
plt.show()
```
